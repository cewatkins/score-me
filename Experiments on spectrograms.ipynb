```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt

from scipy import signal
from scipy.io import wavfile
```

## Upload the .wav file

```python
# I have stereo wav file, its size is 40 Mb, I don't want to upload it to the repository
# You can use your own file and just specify its name
wav_dir = './'
wav_file_name = 'Talk_tonight.wav'

sample_rate, samples = wavfile.read(wav_dir + wav_file_name)
```

```python
# Use samples only from one channel 
samples = np.array[x[0] for x in samples]
```

## Build the spectogram

`nperseg` - length of segment to which the FFT is applied (in samples). The higher it is, the better is resolution in the frequency domain and the worse is resolution in the time domain. Of course, the opposite also holds.

```python
frequencies, times, spectrogram = signal.spectrogram(samples, sample_rate, nperseg=256)

# Diminish the range of amplitudes
spectrogram = np.log(spectrogram)
```

## Plot the spectogram

```python
fig, ax = plt.subplots()

mesh_map = ax.pcolormesh(times, frequencies, spectrogram)
ax.set_ylabel('Frequency [Hz]')
ax.set_xlabel('Time [sec]')
fig.colorbar(mesh_map)
```
